````python
import numpy as np
import matplotlib.pyplot as plt

def ShowVectors(*Vectors):
    # Î≤°ÌÑ∞ Ï†ïÏùò
    # ÏõêÏ†ê
    origin = np.array([0, 0])

    plt.figure(figsize=(6, 6))
    colors=['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
    for i,ve in enumerate(Vectors):
        plt.quiver(
            origin[0], origin[1],       # ÏãúÏûëÏ†ê x, y (Ïó¨Í∏∞ÏÑ† Î™®Îëê ÏõêÏ†ê)
            ve[0], ve[1],               # v1Ïùò x, y ÏÑ±Î∂Ñ
            angles='xy', scale_units='xy', scale=1, color=colors[i], label='v'+f'{i}'
        )
    # Ï¢åÌëú Î≤îÏúÑÏôÄ ÎààÍ∏à ÏÑ§Ï†ï
    plt.xlim(-4, 10)
    plt.ylim(-4, 10)
    plt.xticks(range(-4, 10, 1))    # Í≤©ÏûêÏÑ§Ï†ï
    plt.yticks(range(-4, 10, 1))    # 
    plt.grid(True)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.show()
````
